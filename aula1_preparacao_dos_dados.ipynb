```python
#SUBSTITUIÇÃO DE VALORES AUSENTES

data = pd.read_csv('iris-with-errors.csv', header=(0))
print(data.shape)
data.head(25)
```

```python
#mudando as entradas '?' para 'NaN'

import numpy as np
data = data.replace('?', np.nan)
data.head(25)
```

```python
#substituindo 'NaN' pela média de cada atributo

#convertendo para o formato Numpy e ignoramos a última coluna 
X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
#calculamos a média de cada atributo e armazenamos em um vetor
#usamos a função nanmean que calcula a média ignorando os NaN
averages = np.nanmean(X, axis=0)

for i in np.arange(0, X.shape[0]):
    for j in np.arange(0, X.shape[1]):
    #verificando se uma entrada é NaN
        if np.isnan(X[i,j]) == True:
          #insere a média
          X[i,j] = averages[j]

print(X)
```

```python
#NORMALIZAÇÃO E PADRONIZAÇÃO

import pandas as pd

data = pd.read_csv('iris.csv', header=(0))
print(data.shape)
data.head(25)
```

```python
#normalizando os dados no intervalo [0,1]
import numpy as np
from sklearn.preprocessing import MinMaxScaler
#considera somentes os atributos, ignorando a última coluna, que contem a classe
X = np.array(data[data.columns[0:data.shape[1]-1]])
for i in range(X.shape[1]):
    print("Maior valor da coluna:", i, ":", max(X[:,i]))
    print("Menor valor da coluna:", i, ":", min(X[:,i]))

#prepara a função para transformar os dados
scaler = MinMaxScaler(feature_range=(0, 1))
#realiza a normalização e coloca em um novo vetor
X_norm = scaler.fit_transform(X)
print(X_norm)
```

```python
print('Mínimo dos atributos:', np.amin(X_norm, axis=0))
print('Máximo dos atributos:', np.amax(X_norm, axis=0))
```

```python
#Padronizando os dados de forma que os atributos apresentem média igual a zero e variância igual a 1
from sklearn.preprocessing import StandardScaler
X = np.array(data[data.columns[0:data.shape[1]-1]])
#prepara a função para transformar os dados 
scaler = StandardScaler().fit(X)
#realiza a padronização (média=0, variância=1)
rescalex = scaler.transform(X)
print(rescalex)
```

```python
for i in np.arange(0, rescalex.shape[1]):
    print('Média da coluna', i, 'é igual a: ', np.mean(rescalex[:,i]))
    print('Desvio padrão da coluna', i, 'é igual a: ', np.std(rescalex[:,i]), '\n')
```

```python
#BINARIZAÇÃO DOS DADOS (resulta em valores zero ou um, de acordo com um dado limiar)

from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import MinMaxScaler

X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
T = 0.2 #valor limiar
print('Limiar:', T)
print('--------------------')
      
#prepara a normalização dos dados
scaler = MinMaxScaler(feature_range=(0,1))
#realiza a normalização
X_pad = scaler.fit_transform(X)
#faz a binarização considerando o limiar (threshold)
binarizer = Binarizer(threshold=T).fit(X_pad)
binaryX = binarizer.transform(X_pad)
for i in np.arange(0,X_pad.shape[0]):
    print("Antes:", X_pad[i,])
    print('Depois:', binaryX[i,])
    print('--------------------')
```

```python
#DADOS CORRELACIONADOS(vamos identificar as colunas em que a correlação entre elas é maior do que |p|
data = pd.read_csv('BostonHousing.csv', header=(0))
data.head(10)
```

```python
#obtem a matriz de correlação dos dados
corr = data.corr()
#Mostra a matriz
plt.figure(figsize=(7,7))
#imshow é usado para mostrar imagens
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
#mostra a barra lateral de cores
plt.colorbar()
#inclui o nome das variáveis
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
#os dados estão armazenados na matriz
print(corr)
```

```python
p = 0.75 #correlação mínima
var = []
for i in corr.columns:
    for j in corr.columns:
        if i != j:
            if np.abs(corr[i][j]) > p: #se maior do que |p|
                var.append([i,j])
print('Variáveis mais correlacionadas:\n', var)
```

```python
#DADOS DESBALANCEADOS
data = pd.read_csv('Vehicle.csv', header=(0))
data.head(10)
```

```python
classes = data[data.columns[-1]]
print(classes)
```

```python
#histograma do número de elementos em cada classe

#armazena os nomes das classes
cl = np.unique(classes)
#armazena o número de elementos em cada classe
ncl = np.zeros(len(cl))
for i in np.arange(0, len(cl)):
    a = classes == cl[i]
    ncl[i] = len(classes[a])
print(ncl)    
```

```python
#Em um gráfico de barras

#número de classes
numbers = np.arange(0, len(cl))
plt.bar(numbers, ncl, alpha=.75)
#mostra o nome das classes ao invés dos números
plt.xticks(numbers, cl)
plt.title('Número de elementos em cada classe')
plt.show(True)
```

```python
#OUTLIERS - BOX PLOT
#Em muitos casos, temos a presença de outliers nos dados. 
#Uma maneira simples de identificar outliers é criando um boxplot.

# biblioteca para mostrar os gráficos
import seaborn as sns
data = pd.read_csv('iris.csv', header=(0))
plt.figure(figsize=(8, 8))
# mostra o boxplot
sns.boxplot(x="species", y="petal_length", data=data)
plt.xlabel('Espécie', fontsize=18)
plt.ylabel('Comprimento da pétala', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
#Podemos ainda usar a distância interquantil (Q3 - Q1) para encontrar os outliers
#de uma forma simples. Vamos considerar um exemplo.
np.random.seed(101)
# geramos os dados de forma aleatória e colocamos em um dataframe
data = pd.DataFrame({'a': np.random.randint(1, 200, 20), 
                          'b': np.random.randint(1, 200, 20),
                          'c': np.random.randint(1, 200, 20)})

# geramos os outliers, aumentando em 10 vezes o valor dos atributos
data[data > 150] = data[data > 150]*10
# mostramos os dados
print (data)
```

```python
#Visualizando os dados
import seaborn as sns
sns.pairplot(data)
plt.show()
```

```python
#Para encontrarmos os outliers
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))))
```

```python
#Podemos dizer que uma observação é um outlier se ao menos uma das variáveis está fora dos
#limites máximos do boxplot. Ou seja, se o valor é menor do que (Q1 - 1.5 IQR) ou maior do 
#que (Q3 + 1.5 IQR). Outliers são indicados como True abaixo.
print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)) #is used to invert the answer 0     False
```

```python
#A partir dessa informação, podemos fazer a retirada dos outliers no dados originais.
v = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)
#v = np.invert(v)
data = data.drop(data.index[list(v)], axis=0)
data.head(25)
```
