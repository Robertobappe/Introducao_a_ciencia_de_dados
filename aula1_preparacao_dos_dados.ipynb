```python
#substituindo 'NaN' pela mÃ©dia de cada atributo

#convertendo para o formato Numpy e ignoramos a Ãºltima coluna 
X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
#calculamos a mÃ©dia de cada atributo e armazenamos em um vetor
#usamos a funÃ§Ã£o nanmean que calcula a mÃ©dia ignorando os NaN
averages = np.nanmean(X, axis=0)

for i in np.arange(0, X.shape[0]):
    for j in np.arange(0, X.shape[1]):
    #verificando se uma entrada Ã© NaN
        if np.isnan(X[i,j]) == True:
          #insere a mÃ©dia
          X[i,j] = averages[j]

print(X)
```

```python
#normalizando os dados no intervalo [0,1]
import numpy as np
from sklearn.preprocessing import MinMaxScaler
#considera somentes os atributos, ignorando a Ãºltima coluna, que contem a classe
X = np.array(data[data.columns[0:data.shape[1]-1]])
for i in range(X.shape[1]):
    print("Maior valor da coluna:", i, ":", max(X[:,i]))
    print("Menor valor da coluna:", i, ":", min(X[:,i]))

#prepara a funÃ§Ã£o para transformar os dados
scaler = MinMaxScaler(feature_range=(0, 1))
#realiza a normalizaÃ§Ã£o e coloca em um novo vetor
X_norm = scaler.fit_transform(X)
print(X_norm)
```

```python
print('MÃ­nimo dos atributos:', np.amin(X_norm, axis=0))
print('MÃ¡ximo dos atributos:', np.amax(X_norm, axis=0))
```

```python
#Padronizando os dados de forma que os atributos apresentem mÃ©dia igual a zero e variÃ¢ncia igual a 1
from sklearn.preprocessing import StandardScaler
X = np.array(data[data.columns[0:data.shape[1]-1]])
#prepara a funÃ§Ã£o para transformar os dados 
scaler = StandardScaler().fit(X)
#realiza a padronizaÃ§Ã£o (mÃ©dia=0, variÃ¢ncia=1)
rescalex = scaler.transform(X)
print(rescalex)
```

```python
for i in np.arange(0, rescalex.shape[1]):
    print('MÃ©dia da coluna', i, 'Ã© igual a: ', np.mean(rescalex[:,i]))
    print('Desvio padrÃ£o da coluna', i, 'Ã© igual a: ', np.std(rescalex[:,i]), '\n')
```

```python
#BINARIZAÃÃO DOS DADOS (resulta em valores zero ou um, de acordo com um dado limiar)

from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import MinMaxScaler

X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
T = 0.2 #valor limiar
print('Limiar:', T)
print('--------------------')
      
#prepara a normalizaÃ§Ã£o dos dados
scaler = MinMaxScaler(feature_range=(0,1))
#realiza a normalizaÃ§Ã£o
X_pad = scaler.fit_transform(X)
#faz a binarizaÃ§Ã£o considerando o limiar (threshold)
binarizer = Binarizer(threshold=T).fit(X_pad)
binaryX = binarizer.transform(X_pad)
for i in np.arange(0,X_pad.shape[0]):
    print("Antes:", X_pad[i,])
    print('Depois:', binaryX[i,])
    print('--------------------')
```

```python
#DADOS CORRELACIONADOS(vamos identificar as colunas em que a correlaÃ§Ã£o entre elas Ã© maior do que |p|
data = pd.read_csv('BostonHousing.csv', header=(0))
data.head(10)
```

```python
#obtem a matriz de correlaÃ§Ã£o dos dados
corr = data.corr()
#Mostra a matriz
plt.figure(figsize=(7,7))
#imshow Ã© usado para mostrar imagens
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
#mostra a barra lateral de cores
plt.colorbar()
#inclui o nome das variÃ¡veis
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
#os dados estÃ£o armazenados na matriz
print(corr)
```

```python
p = 0.75 #correlaÃ§Ã£o mÃ­nima
var = []
for i in corr.columns:
    for j in corr.columns:
        if i != j:
            if np.abs(corr[i][j]) > p: #se maior do que |p|
                var.append([i,j])
print('VariÃ¡veis mais correlacionadas:\n', var)
```

```python
#OUTLIERS - BOX PLOT
#Em muitos casos, temos a presenÃ§a de outliers nos dados. 
#Uma maneira simples de identificar outliers Ã© criando um boxplot.

# biblioteca para mostrar os grÃ¡ficos
import seaborn as sns
data = pd.read_csv('iris.csv', header=(0))
plt.figure(figsize=(8, 8))
# mostra o boxplot
sns.boxplot(x="species", y="petal_length", data=data)
plt.xlabel('EspÃ©cie', fontsize=18)
plt.ylabel('Comprimento da pÃ©tala', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
#Podemos ainda usar a distÃ¢ncia interquantil (Q3 - Q1) para encontrar os outliers
#de uma forma simples. Vamos considerar um exemplo.
np.random.seed(101)
# geramos os dados de forma aleatÃ³ria e colocamos em um dataframe
data = pd.DataFrame({'a': np.random.randint(1, 200, 20), 
                          'b': np.random.randint(1, 200, 20),
                          'c': np.random.randint(1, 200, 20)})

# geramos os outliers, aumentando em 10 vezes o valor dos atributos
data[data > 150] = data[data > 150]*10
# mostramos os dados
print (data)
```

```python
#Visualizando os dados
import seaborn as sns
sns.pairplot(data)
plt.show()
```

```python
#Para encontrarmos os outliers
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))))
```

```python
#Podemos dizer que uma observaÃ§Ã£o Ã© um outlier se ao menos uma das variÃ¡veis estÃ¡ fora dos
#limites mÃ¡ximos do boxplot. Ou seja, se o valor Ã© menor do que (Q1 - 1.5 IQR) ou maior do 
#que (Q3 + 1.5 IQR). Outliers sÃ£o indicados como True abaixo.
print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)) #is used to invert the answer 0     False
```

```python
#A partir dessa informaÃ§Ã£o, podemos fazer a retirada dos outliers no dados originais.
v = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)
#v = np.invert(v)
data = data.drop(data.index[list(v)], axis=0)
data.head(25)
```
