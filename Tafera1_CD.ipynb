```python
#2 - Leia os dados com erro novamente e insira a mediana de cada atributo onde for encontrada 
#NaN
```

```python
data = pd.read_csv('iris-with-errors.csv', header=(0))
data.head(25)
```

```python
#vamos remover as linhas duplicadas
data = data.drop_duplicates() #remove as linhas duplicadas
data.head(25)
```

```python
#Antes de inserimos a mediana vamos substituir '?' por 'NaN'
data = data.replace('?', np.nan)
data.head(25)
```

```python
#vamos aqui inserir a mediana
#conversão para o formato Numpy e ignoramos a última coluna
X = np.array(data[data.columns[0:data.shape[1]-1]], dtype=float)
#calculamos a mediana de cada atributo e armazenamos em um vetor
#usamos a função nanmedian que clacula a mediana ignorando NaN
medians = np.nanmedian(X, axis=0)
for i in np.arange(0, X.shape[0]):
    for j in np.arange(0, X.shape[1]):
        if np.isnan(X[i,j]) == True: #verificando se é uma entrada NaN
            X[i,j] = medians[j] #insera a mediana
print(X)
```

```python
#3 - Conside os dados da Iris e mostre a distribuição de probabilidades de cada uma das variáveis 
#após a normalização e padronização
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
#Gerando o histogrando antes da normalização e padronização
X = np.array(data[data.columns[0:data.shape[1]-1]])

#Gerando a distribuição
import matplotlib.pyplot as plt
plt.hist(X, bins=10)
plt.xlabel('Valores')
plt.ylabel('Contagem')
```

Notamos que os valores estão entre (0,8)

```python
#vamos normalizar os dados no intervalo [0,1]
from sklearn.preprocessing import MinMaxScaler

X = np.array(data[data.columns[0:data.shape[1]-1]])
for i in range(X.shape[1]):
    print("Maior valor da coluna:", i,":", max(X[:,i]))
    print("Menor valor da coluna:", i,":", min(X[:,i]))
    
#prepara a função para transformar os dados
scaler = MinMaxScaler(feature_range=(0,1))
#realiza a normalização e coloca em um novo vetor
X_norm = scaler.fit_transform(X)
print(X_norm)
```

```python
print('Mínimo dos atributos:', np.amin(X_norm, axis=0))
print('Máximo dos atributos:', np.amax(X_norm, axis=0))
```

```python
#Gerando a distribuição
import matplotlib.pyplot as plt
plt.hist(X_norm, bins=10)
plt.xlabel('Valores normalizados')
plt.ylabel('Contagem')
```

Após a normalização os valores agora estão entre (0,1)

```python
#vamos padronizar os dados de modo que os atributos apresentem média igual a zero
#e variância igual a 1

from sklearn.preprocessing import StandardScaler
X = np.array(data[data.columns[0:data.shape[1]-1]])
#prepara a função para transformar os dados
scaler = StandardScaler().fit(X)
#realiza a padronização(média=0, variância=1)
rescaledX = scaler.transform(X)
print(rescaledX)
```

```python
#verificando se a média é zero e a variância igual a 1
for i in np.arange(0, rescaledX.shape[1]):
    print('Média da coluna', i, 'é igual a: ', np.mean(rescaledX[:,i]))
    print('Desvio padrão da coluna', i, 'é igual a: ', np.std(rescaledX[:,i]), '\n')
```

```python
plt.hist(rescaledX, bins=10)
plt.xlabel('Valores padronizados')
plt.ylabel('Contagem')
```

Após a padronização temos agora também valores negativos com um pouco de assimetria sendo a cursa similar a uma gaussiana. Apesar desta assimetria, notamos média próxima de zero.

Ambas as técnicas tem como objetivo deixar as variáveis na mesma ordem de grandeza. Enquanto na normalização transforma os dados num intervalo entre 0 e 1 (ou -1 e 1, caso tenha valores negativos), a padronização transforma os dados de modo que a média igual 0 e desvio padrão 1. A importância de deixar os dados na mesma escala é para pode aplicar alguns algoritmos de machine learning.  

```python
#4 Reamostre os dados da Iris e selecione 10 elementos em cada classe
#5 MOnte um boxplot para cada variável dos dados da iris
data = pd.read_csv('iris.csv')
data.head(25)
```

```python
data[['sepal_length', 'species']][0:10]
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))

sns.boxplot(x="species", y="sepal_length", data=data[['sepal_length', 'species']][0:10]) #classe sepal_length
plt.xlabel('Espécie', fontsize=18)
plt.ylabel("Comprimento da pétala", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show(True)
```

```python
#encontrando os outliers

#vamos usar a distância interquantil (Q3 - Q1) para encontrar os outliers
Q1 = data[['sepal_length', 'species']][0:10].quantile(0.25)
Q3 = data[['sepal_length', 'species']][0:10].quantile(0.75)
IQR = Q3 - Q1

print (((data[['sepal_length', 'species']][0:10] < (Q1 - 1.5 * IQR)) | 
        (data[['sepal_length', 'species']][0:10] > (Q3 + 1.5 * IQR))))
```

```python
print (((data[['sepal_length', 'species']][0:10] < (Q1 - 1.5 * IQR)) | 
        (data[['sepal_length', 'species']][0:10] > (Q3 + 1.5 * IQR))).any(axis=1))
```
