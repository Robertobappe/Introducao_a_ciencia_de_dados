```python
#1 - Gere dados a partir de uma distribuição de Poisson. Varie a taxas �λ no intervalo [1,10] e 
#mostre o gráfico da média em função da variância.
```

```python
import pandas as pd
import numpy as np

X, Y = [], []

for k in range(1,11):
    data = np.random.poisson(lam=k, size=100)
    y = np.mean(data)
    Y.append(y)
    x = np.var(data)
    X.append(x)

print(X)
print(Y)
```

```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(6,4))

plt.scatter(X, Y, marker='o', color = 'black');
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Variância",fontsize = 15) 
plt.ylabel("Média", fontsize = 15) 
plt.show(True)
```

```python
#4 - Para a função log(), investigue como as correlações de Pearson e Spearman variam de acordo 
#com o intervalo dos dados.
```

```python
from scipy.stats import pearsonr, spearmanr

for N in range(100, 1_100, 100):
    x = np.linspace(1, 50000, N)
    z = np.log(x)
    plt.plot(x, z) 
    plt.xlabel("x",fontsize = 20) 
    plt.ylabel("Log(X)", fontsize = 20) 
    corr, p_value = pearsonr(x, z)
    corrs, p_values = spearmanr(x, z)
    corr = int(corr*100)/100
    corrs = int(corrs*100)/100
    string  = 'Value N = '+ str(N)
    plt.text(20000,3, string, fontsize=15)
    string = 'Pearson corr. = '+ str(corr)
    plt.text(20000,2, string, fontsize=15)
    string = 'Spearman corr. = '+ str(corrs)
    plt.text(20000,1, string, fontsize=15)
    plt.show()
```

```python
#observamos que quando o valor de N aumenta o coorelação de Pearson vai aumentando de valor de 0.72
#até 0.84 enquanto que a coorelação de Spearman fica entre [0.99, 1]. 
```

```python
#5 - Considere o código acima que mostra como a correlação de Pearson muda com a inclusão de 
#ruídos. Modifique a função para Y=0.5∗X+Y=0.5∗X+ruído. Varie o ruído e calcule os 
#coeficientes de Pearson e Spearman, mostrando os respectivos scatterplots com os valores dos 
#coeficientes (como feito no exemplo).

#usando o código do exemplo
from scipy.stats import pearsonr

N = 100
X = np.linspace(-1,1, N)
erro = np.random.uniform(-1,1,N) # ruído a ser incluído na relação linear.
for sigma in np.arange(0,2,0.2):
    Y = -0.8*X + erro*sigma
    plt.plot(X,Y, 'ro')
    corr, p_value = pearsonr(X, Y) # calcula a correlação
    corr = int(corr*100)/100
    string = 'corr = '+ str(corr)
    plt.xlim(-1.5,1.5)
    plt.ylim(-2, 2)
    plt.text(0.6,1.7, string, fontsize=15)
    plt.show(True)
```

```python
#vamos modificar a função para Y=0.5*X+Y = 0.5*X+ruído
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

N = 100
X = np.linspace(-1,1, N)
erro = np.random.uniform(-1,1,N) # ruído a ser incluído na relação linear.
for sigma in np.arange(0,2,0.2):
    Y = 0.5*X + erro*sigma
    plt.plot(X,Y, 'ro')
    corr, p_value = pearsonr(X, Y) # calcula a correlação de Pearsonr
    corrs, p_values = spearmanr(X, Y) # calcula a correlação de Spearmanr
    corr = int(corr*100)/100
    string = 'P corr = '+ str(corr)
    corrs = int(corrs*100)/100
    strings = 'S corr = '+ str(corrs)
    plt.xlim(-1.5,1.5)
    plt.ylim(-2, 2)
    plt.text(0.6,1.7, string, fontsize=15)
    plt.text(0.6,1.2, strings, fontsize=15)
    plt.show(True)
```
